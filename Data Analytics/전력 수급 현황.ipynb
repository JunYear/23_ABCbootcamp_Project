# 가구 평균 월별 전력사용량 - 서울


```python
household_power_usage_seoul_df = pd.read_csv('/content/drive/MyDrive/abc부트캠프 proj_2조/준/가구 평균 전력 사용량 (서울).csv', encoding='UTF-8')
household_power_usage_seoul_df
```

## 년도별 평균으로 전처리

```python
# 년도별로 데이터를 그룹화하고 평균을 계산하는 함수
def compute_yearly_mean(df, year_mapping):
    result = {}
    for year, columns in year_mapping.items():
        result[year] = df[columns].mean(axis=1)
    return pd.DataFrame(result)

# 년도별로 컬럼 그룹화
year_grouped_columns = {}
for col in household_power_usage_seoul_df.columns[1:]:
    year = col.split("년")[0]
    year_grouped_columns.setdefault(year, []).append(col)

# 년도별 평균 계산
household_power_usage_seoul_yr_df = compute_yearly_mean(household_power_usage_seoul_df, year_grouped_columns)

# '시군구' 컬럼을 추가
household_power_usage_seoul_yr_df['시군구'] = household_power_usage_seoul_df['시군구']

# 결과 확인
household_power_usage_seoul_yr_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(24, 14))  # 크기를 더 크게 조정
sns.lineplot(data=household_power_usage_seoul_yr_df, x='시군구', y='13', marker="o", label="2013")
sns.lineplot(data=household_power_usage_seoul_yr_df, x='시군구', y='14', marker="o", label="2014")
sns.lineplot(data=household_power_usage_seoul_yr_df, x='시군구', y='21', marker="o", label="2021")
sns.lineplot(data=household_power_usage_seoul_yr_df, x='시군구', y='22', marker="o", label="2022")

plt.xlabel('시군구', fontsize=30)  # 폰트 크기 조정
plt.ylabel('전력량', fontsize=30)  # 폰트 크기 조정
plt.xticks(rotation=45, fontsize=20)  # 폰트 크기 조정
plt.yticks(fontsize=30)  # 폰트 크기 조정
plt.grid(True, axis='y', linestyle='--', linewidth=0.5)
plt.legend(title='년도', fontsize=30, title_fontsize=20)  # 폰트 크기 조정
plt.tight_layout()
plt.show()
```
