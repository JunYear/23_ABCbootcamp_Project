# 온실가스와 이산화탄소의 연도별 총배출량 그래프
출처 :https://www.index.go.kr/unity/potal/indicator/ProgressIndicator.do?cdNo=210

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/차지예/온시가스,CO2 연도별 통계1990_2020.csv', encoding='euc-kr')
data
```

```python
import matplotlib.pyplot as plt

# Extracting the data for "온실가스 총배출량" and "CO2 총배출량"
ghg_emission = data.iloc[0, 1:].str.replace(',', '').astype(float)
co2_emission = data.iloc[2, 1:].str.replace(',', '').astype(float)
years = data.columns[1:]

# Plotting the data
plt.figure(figsize=(15,7))
plt.plot(years, ghg_emission, label='온실가스 총배출량', marker='o', linestyle='-')
plt.plot(years, co2_emission, label='CO2 총배출량', marker='o', linestyle='--')
plt.xticks(rotation=45)
plt.title('온실가스와 CO2의 연도별 총배출량 (1990-2020)')
plt.xlabel('연도')
plt.ylabel('배출량 (단위: Gg)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
온실가스와 CO2의 연도별 총배출량
```

##1차사업소명별 연도별 전력사용량
출처:https://epsis.kpx.or.kr/epsisnew/selectMain.do?locale=

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/차지예/한국전력공사_연도별 전력사용량_20221231.csv", encoding='cp949')
```

```python
# 데이터를 사업소명과 연도로 피벗
pivot_data_transposed = data.pivot(index='1차사업소명', columns='연도', values='전력사용량')

# 각 연도별로 전력 사용량을 퍼센트로 변환 (퇴적 100% 차트를 위해)
pivot_data_percentage = pivot_data_transposed.divide(pivot_data_transposed.sum()) * 100

# 퇴적 100% 영역 차트 그리기
ax = pivot_data_percentage.plot(kind='area', stacked=True, figsize=(12, 7), colormap='viridis')
ax.set_title('사업소별 연도별 전력 사용량 퇴적 100% 차트')
ax.set_xlabel('사업소명')
ax.set_ylabel('전력 사용량 (%)')
ax.grid(axis='y', linestyle='--', linewidth=0.5)
ax.set_xticks(range(len(pivot_data_percentage)))
ax.set_xticklabels(pivot_data_percentage.index, rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 퇴적 영역 차트 그리기
ax = pivot_data_transposed.plot(kind='area', figsize=(12, 7), colormap='viridis', alpha=0.7)
ax.set_title('사업소별 연도별 전력 사용량 퇴적 영역 차트')
ax.set_xlabel('사업소명')
ax.set_ylabel('전력 사용량 (단위: kWh)')
ax.grid(axis='y', linestyle='--', linewidth=0.5)
ax.set_xticks(range(len(pivot_data_transposed)))
ax.set_xticklabels(pivot_data_transposed.index, rotation=45)

plt.tight_layout()
plt.show()
```

## 이상기후 ->연도별 열대야,폭염 빈도
출처 :https://www.weather.go.kr/w/index.do

```python
import matplotlib.pyplot as plt
import pandas as pd

# 데이터 불러오기
tropical_night_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/차지예/열대야1991_2023.csv", encoding='cp949')
heatwave_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/차지예/폭염1991_2023.csv", encoding='cp949')

# '연도'와 '연합계' 컬럼만 추출
tropical_night_sum = tropical_night_data[['연도', '연합계']]
heatwave_sum = heatwave_data[['연도', '연합계']]

# 평년 데이터 제거
tropical_night_sum = tropical_night_sum[tropical_night_sum['연도'] != '평년']
heatwave_sum = heatwave_sum[heatwave_sum['연도'] != '평년']

# 데이터 타입 변환
tropical_night_sum['연도'] = tropical_night_sum['연도'].astype(int)
heatwave_sum['연도'] = heatwave_sum['연도'].astype(int)

bar_height = 0.4

# 항아리 모양의 그래프 그리기
plt.figure(figsize=(10, 7))
index = range(len(tropical_night_sum['연도']))

# 열대야 데이터는 양수로, 폭염 데이터는 음수로 표시
bar1 = plt.barh(index, tropical_night_sum['연합계'], bar_height, label='열대야', color='blue', align='center')
bar2 = plt.barh([i + bar_height for i in index], -heatwave_sum['연합계'], bar_height, label='폭염', color='red', align='center')

plt.ylabel('연도')
plt.xlabel('일수')
plt.title('연도별 열대야와 폭염의 연합계 비교')
plt.yticks([i + bar_height/2 for i in index], tropical_night_sum['연도'])
plt.legend(loc='upper right')
plt.tight_layout()

plt.show()
# 증가 추세: 최근 몇 년 동안 열대야와 폭염의 일수가 전반적으로 증가하고 있는 추세, 열대야와 폭염의 일수가 두드러지게 높은 것을 확인할 수 있습니다.

```

```python
import pandas as pd

# 파일 읽기
data_1973 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/차지예/1973_전국_기온.csv', encoding='cp949')
data_2022 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/차지예/2022_전국_기온.csv', encoding='cp949')

# 두 데이터 프레임의 상위 5개 행을 출력
data_1973.head(), data_2022.head()
```

```python
# 두 데이터프레임을 지점정보를 기준으로 병합 (inner join)
merged_data = pd.merge(data_1973, data_2022, on="지점정보", how="inner", suffixes=('_1973', '_2022'))

# 필요한 컬럼만 선택
result_data = merged_data[["지점정보", "평균기온(°C)_1973", "최저기온(°C)_1973", "최고기온(°C)_1973",
                           "평균기온(°C)_2022", "최저기온(°C)_2022", "최고기온(°C)_2022"]]

result_data.head()
```

```python
# 결과 데이터를 CSV 파일로 저장
file_path = "merged_temperature_data.csv"
result_data.to_csv(file_path, encoding='cp949', index=False)

file_path
```

```python
# 파일 읽기
merged_data = pd.read_csv('/content/merged_temperature_data.csv', encoding='cp949')

# 1973년과 2022년의 평균기온, 최저기온, 최고기온의 합계를 계산
avg_temp_diff = merged_data["평균기온(°C)_2022"].sum() - merged_data["평균기온(°C)_1973"].sum()
min_temp_diff = merged_data["최저기온(°C)_2022"].sum() - merged_data["최저기온(°C)_1973"].sum()
max_temp_diff = merged_data["최고기온(°C)_2022"].sum() - merged_data["최고기온(°C)_1973"].sum()

avg_temp_diff, min_temp_diff, max_temp_diff
```

```python
import matplotlib.pyplot as plt

def draw_half_donut_chart(title, value_1973, value_2022, ax):
    # 데이터 설정
    data = [value_1973, value_2022 - value_1973]
    colors = ['#FF5733', '#E0E0E0']  # 1973년 데이터의 색상, 나머지의 색상
    wedges, texts, autotexts = ax.pie(data, colors=colors, startangle=90, counterclock=False,
                                      wedgeprops=dict(width=0.4), autopct='%1.1f%%', pctdistance=0.85)

    # 반원만 보이게 설정
    ax.set_aspect("equal")
    ax.add_artist(plt.Circle((0,0),0.70,fc='white'))

    # 제목 설정
    ax.set_title(title, y=0.5)

# 그림 생성
fig, axs = plt.subplots(1, 3, figsize=(15,5))

draw_half_donut_chart("평균기온", merged_data["평균기온(°C)_1973"].sum(), merged_data["평균기온(°C)_2022"].sum(), axs[0])
draw_half_donut_chart("최저기온", merged_data["최저기온(°C)_1973"].sum(), merged_data["최저기온(°C)_2022"].sum(), axs[1])
draw_half_donut_chart("최고기온", merged_data["최고기온(°C)_1973"].sum(), merged_data["최고기온(°C)_2022"].sum(), axs[2])

plt.tight_layout()
plt.show()
#평균기온은 1973년 보다 2022년이 6.2% 기온이 올랐다 : 50도차이
#최저기온은 1973년 보다 2022년이 8.8% 기온이 올랐다 : 45도차이
#최고기온은 1973년 보다 2022년이 5.4% 기온이 올랐다 : 61도차이
```
